```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("..\kc_house_data.csv")
pd.set_option('precision',10)
```

```python
df.head()
```

## 1- Qual o número de imóveis por ano de construção?

```python
df[['id', 'yr_built']].groupby(['yr_built']).count().reset_index()
```

## 2- Qual o menor número de quartos por ano de construção de imóveis?

```python
df[['bedrooms', 'yr_built']].groupby(['yr_built']).min().reset_index()
```

## 3- Qual o preço de compra mais alto por cada número de quarto?

```python
df[['price', 'bedrooms']].groupby(['bedrooms']).max().reset_index()
```

## 4- Qual a soma de todos os preços de compra por número de quartos?

```python
df[['price', 'bedrooms']].groupby('bedrooms').sum().reset_index()
```

## 5- Qual a soma de todos os preços de compra por número de quartos e banheiros?

```python
df[['price', 'bedrooms', 'bathrooms']].groupby(['bathrooms', 'bedrooms']).sum().reset_index()
```

## 6- Qual o tamanho médio das salas dos imóveis por ano de construção?

```python
pd.set_option('precision', 2)
df[['sqft_living', 'yr_built']].groupby(['yr_built']).mean().reset_index()
```

## 7- Qual o tamanho mediano das salas dos imóveis por ano de construção?

```python
df[['sqft_living', 'yr_built']].groupby(['yr_built']).median().reset_index()
```

## 8- Qual o desvio-padrão do tamanho das salas dos imóveis por ano de construção?

```python
df[['sqft_living', 'yr_built']].groupby(['yr_built']).std().reset_index()
```

## 9- Como é o crescimento médio de preços de compras dos imóveis: por ano, por dia e pela semana do ano?

```python
df['year'] = pd.to_datetime(df['date']).dt.year

data = df[['price', 'year']].groupby('year').sum().reset_index()

plt.bar(data['year'], data['price'])
```

```python
df['day'] = pd.to_datetime(df['date'])

by_date = df[['price', 'day']].groupby('day').mean().reset_index()

plt.figure(figsize = (8,8))
plt.plot(by_date['day'], by_date['price'])
```

```python
df['year_week'] = pd.to_datetime(df['date']).dt.strftime('%Y-%W')

by_week = df[['price', 'year_week']].groupby('year_week').mean().reset_index()

plt.figure(figsize = (15,15))
plt.bar(by_week['year_week'], by_week['price'])
plt.xticks(rotation = 60);
```

## Criação de um Dashboard

```python
from matplotlib import gridspec

fig = plt.figure(figsize = (20, 12))
specs = gridspec.GridSpec(ncols = 2, nrows = 2, figure = fig)

ax1 = fig.add_subplot(specs[0, :]) #Primeira linha inteira
ax2 = fig.add_subplot(specs[1, 0]) #Segunda linha, primeira coluna
ax3 = fig.add_subplot(specs[1, 1]) #Segunda linha, segunda coluna

#Primeiro gráfico
df['year'] = pd.to_datetime(df['date']).dt.year
data = df[['price', 'year']].groupby('year').sum().reset_index()
ax1.bar(data['year'], data['price'])

#Segundo gráfico
df['day'] = pd.to_datetime(df['date'])
by_date = df[['price', 'day']].groupby('day').mean().reset_index()
ax2.plot(by_date['day'], by_date['price'])

#Terceiro gráfico
df['year_week'] = pd.to_datetime(df['date']).dt.strftime('%Y-%W')
by_week = df[['price', 'year_week']].groupby('year_week').mean().reset_index()
ax3.bar(by_week['year_week'], by_week['price'])
plt.xticks(rotation = 60);
```

## 10- Eu gostaria de olhar no mapa e conseguir identificar as casas com maior preço

```python
import plotly.express as px

data = df[['id', 'lat', 'long', 'price']]
```

```python
fig = px.scatter_mapbox(data, lat = 'lat', lon = 'long',
                        size = 'price',
                        color_continuous_scale = px.colors.cyclical.IceFire,
                        zoom = 10, size_max = 15)

fig.update_layout(mapbox_style = 'open-street-map')
fig.update_layout(height = 600, margin = {'r': 0, 't': 0, 'l': 0, 'b': 0})
fig.show()
```
